# Normalizing data

```python
sc = StandardScaler()
cols=test_dataset.columns
train = sc.fit_transform(train)
test = sc.transform(test_dataset)
train = pd.DataFrame(train,columns=cols)
test = pd.DataFrame(test,columns=cols)
```

```python
train.head()
```

```python
test.head()
```
