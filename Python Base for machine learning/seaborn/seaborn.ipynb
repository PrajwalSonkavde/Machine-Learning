```python
df = sns.load_dataset("iris")
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['sepal_length'],kde=False,bins=10)
```

```python
sns.countplot('species',data=df)
```

```python
sns.countplot(y='species',data=df)
```

```python
sns.barplot(x='sepal_length',y='species',data=df)
```

```python
sns.boxplot('species','sepal_length',data=df,palette='rainbow')
```
